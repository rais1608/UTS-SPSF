```python
import numpy as np

# Masukkan nilai resistor dan tegangan sebagai input
R1 = 1
R2 = 2
R3 = 3
R4 = 4
R5 = 5
R6 = 6
R7 = 7
R8 = 8

E1 = 12
E2 = 15
E3 = 16
E4 = 19

# Membuat matriks koefisien
A = np.array([[R1+R5+R6, R2, -R6],
              [R5, R2+R5+R8, -R8],
              [-R6, -R8, R3+R6+R7],
              [0, -R8, R7+R8+R4]])

# Membuat vektor hasil
B = np.array([E1, E2, E3, E4])

# Mencari solusi sistem persamaan linear
I = np.linelg.solve(A,B)

# Menampilkan hasil
print("Nilai arus I1: {:.2f}".format(I[0]))
print("Nilai arus I2: {:.2f}".format(I[1]))
print("Nilai arus I3: {:.2f}".format(I[2]))
print("Nilai arus I4: {:.2f}".format(I[3]))

print (I)
```
